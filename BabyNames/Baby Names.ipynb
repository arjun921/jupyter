```python
import sys
import re
""" prints out text without html tags"""
def extract_names(filename):
        open_file = open(filename,'r')
        line = open_file.readline()
        removeList = ["tr","align","right","td","colspan","a","href","table","colspan","p","class","valign","middle", "nbsp", "th", "scope", "col", "width", "bgcolor","cellpadding","cellspacing", "bordercolor","h1","h2", "h3", "html","h4", "h5", "input", "type", "hidden", "submit", "value", "form", "method","background","border"]
        remove = '|'.join(removeList)
        remReg = re.compile(r'\b('+remove+r')\b', flags=re.IGNORECASE)
        while line:
                line = (' '.join(re.split('\W+',line)))
                print(remReg.sub("",line))
                line = open_file.readline()
        return


def main():
        args = sys.argv[1]
        if not args:
                print ('usage: [--summaryfile] file [file ...]')
                sys.exit(1)
        extract_names("baby1990.html")


main()

```
